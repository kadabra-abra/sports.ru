## Импорт данных и библиотек:

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./dataset_Facebook.csv', sep=';')
data.head(10)
```

## Статистики по столбцам:

```python
def statistics(data):
    s1 = data.describe()
    mode = data.mode(axis='rows').head(1).rename({0: 'mode'}, axis='index')
    s2 = pd.concat([s1, mode])
    return(s2)
```

```python
statistics(data)
```

```python
np.unique(data.Type)
```

### Статистики для фотографий:

```python
photos = data[data.Type == 'Photo']
statistics(photos)
```

### Статистики для статусов:

```python
status = data[data.Type == 'Status']
statistics(status)
```

### Статистики для ссылок:

```python
links = data[data.Type == 'Link']
statistics(links)
```

### Статистики для видео:

```python
videos = data[data.Type == 'Video']
statistics(videos)
```
